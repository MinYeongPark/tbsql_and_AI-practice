```python
import cv2
```

```python
img = cv2.imread('/content/sampleimg.jpg') # 경로 복사 붙여넣기 해야 함
```

```python
# 직접 사진 보기
from matplotlib import pyplot as plt
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # # cv2는 gbr임. 그래서 rgb로 바꾸어야 함.
plt.imshow(img)
```

1. 라인 그리기

```python
cv2.line(img, (4,4), (300,300), (0,0,255), 5 ) # 시작점, 끝점, 색상, 두께  # 여러번 실행하면 중첩해서 그려짐
# 이미지가 나타내는 픽셀 값을 출력함
```

```python
plt.imshow(img)
# 선이 그려진 것을 확인할 수 있음.
```

```python
img = cv2.rectangle(img, (100,100), (240,300), (10,250,100), 10 )
plt.imshow(img)
```

```python
text = 'Hello world'
img = cv2.putText(img, text, (50,150), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,255,0), 3 ) # 텍스트, 시작할 위치, 폰트 등등
                                                                                     # 파라미터 순서 기억할 필요 없음. 검색하면 됨.
plt.imshow(img)
```
