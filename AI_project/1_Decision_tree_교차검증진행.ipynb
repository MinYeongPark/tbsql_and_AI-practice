```python
# 전체적인 DT 코드 참고 출처 : https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=winddori2002&logNo=221659080425
# 참고 : https://injo.tistory.com/15
#        http://www.gisdeveloper.co.kr/?p=9932

import pandas as pd
import matplotlib.pyplot as plt

# csv 파일 읽기
data = pd.read_csv('./drive/MyDrive/TABA_AI/dataset/entropy11.csv')
data.head()
```

```python
# 66번칼럼부터 120번 칼럼까지 삭제
data.drop(list(map(str, range(66, 121, 1))), axis=1, inplace=True)

# uuid도 삭제
data.drop(['uuid'], axis=1, inplace=True)

# 맨 처음 col도 삭제
data.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
data
```

```python
from sklearn.tree import DecisionTreeClassifier
import sklearn.model_selection as ms

# 타겟 변수의 형변환
mapping_dict = {'Gafgyt' : 1,
                'Mirai' : 2,
                'Benign' : 3,
                'Dofloo' : 4,
                'Tsunami' : 5,}

after_mapping_label = data['label'].apply(lambda x : mapping_dict[x])
```

```python
from sklearn.model_selection import train_test_split

x = data[data.columns.difference(['label'])] #데이터에서 label을 뺀 값
y = after_mapping_label #label을 각각 위의 코드 처럼 Gafgpt를 1로 하고.. 등등으로 라벨링한거

# train 데이터와 test 데이터 7:3 비율로 분할
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                        test_size = 0.3, random_state = 42) 
```

```python
# DT 객체 생성 및 훈련
dt_clf = DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)

# 예측값 저장
y_pred = dt_clf.predict(x_test)
```

```python
import sklearn.metrics as mt

# 학습결과 평가 
print('Train_Accuracy: ', dt_clf.score(x_train, y_train),'\n')
```

```python
accuracy = mt.accuracy_score(y_test, y_pred)                   # 정확도
recall = mt.recall_score(y_test, y_pred,average='micro')       # 재현율
precision = mt.precision_score(y_test, y_pred,average='micro') # 정밀도
f1_score = mt.f1_score(y_test, y_pred, average='micro')        # F1 스코어 계산 (불균형한 데이터에서 잘 동작하는 평가지표)
matrix = mt.confusion_matrix(y_test, y_pred)                  

print('Accuracy: ', format(accuracy,'.2f'),'\n')
print('Recall: ', format(recall,'.2f'),'\n')
print('Precision: ', format(precision,'.2f'),'\n')
print('F1_score: ', format(f1_score,'.2f'),'\n')
print('Confusion Matrix:','\n', matrix)

#                   예측클래스 1     /   예측클래스 2   / 예측클래스 3 / 예측클래스 4 / 예측클래스 5 / 
# 정답클래스 1   정답1&예측1표본 수   정답1&예측2표본 수        ...
# 정답클래스 2
# 정답클래스 3
# 정답클래스 4
# 정답클래스 5
```

```python
# 교차검증
from sklearn.model_selection import cross_val_score, cross_validate

# cross_val_score : 각 검증의 평가 스코어 도출, 간단히 각 검증의 '평가 점수'만을 보여주는 함수
# x : feature 데이터 세트, y : 레이블 데이터 세트, cv : 교차 검증 폴드 수
scores = cross_val_score(dt_clf, x, y, cv = 5) # 
print(f'각 검증 별 점수 :\n{pd.Series(scores)}\n')
print(f'교차검증 평균 점수 :{scores.mean()}\n')

# cross_validate : 각 검증의 결과 도출, 분할마다 훈련과 테스트에 걸린 시간을 담은 딕셔너리를 반환한다
rst = cross_validate(dt_clf, x, y, cv=5, return_train_score=True) 
print('각 검증의 결과 : \n', pd.DataFrame(rst)) # 학습 시간, 평가 시간, 평가 점수, 훈련 점수 출력
```

```python
# 파라미터 튜닝 (그리드 서치)
# 튜닝을 통해 모델의 과적합을 방지하고 모델 스코어를 올릴 수 있음
# GridSearchCV : 튜닝하고 싶은 파라미터를 집어넣어 튜닝과 교차검증을 함께 진행할 수 있게 도와줌

from sklearn.model_selection import GridSearchCV

# 테스트하고자 하는 파라미터 값들을 사전타입으로 정의
dt_clf = DecisionTreeClassifier(random_state=33)
parameters = {'max_depth': [3, 5, 7],            # 트리 모형의 최대 깊이 
                                                 # (max_depth 값이 커질수록 깊어져 과적합 쉽게 일어날 수 있음)
              'min_samples_split': [3, 5],       # 중간노드에서 분리가 일어나기 위한 샘플의 수
              'splitter': ['best', 'random']}    # 노드 분리 방법으로 'random', 'best'를 사용

grid_dt = GridSearchCV(dt_clf,    # estimator 객체,
                      param_grid = parameters, cv = 5,
                      )

grid_dt.fit(x_train, y_train)

# GridSearchCV 결과를 DataFrame으로 변환
result = pd.DataFrame(grid_dt.cv_results_['params'])
result['mean_test_score'] = grid_dt.cv_results_['mean_test_score'] # mean_test_score는 각 fold 5개의 평균을 의미
print(result)
result.sort_values(by='mean_test_score', ascending=False) 

# 결과 해석 :
# max_depth = 7, min_sample_split = 3, splitter = best 일 때 가장 좋은 성능을 보임
```

```python
# 의사결정트리 시각화를 위한 작업
# 트리 시각화를 위한 export_graphviz 모듈 임포트 (이를 위해 graphviz 별도 설치 필요)
# (jupyter notbook에서 실행 시) graphviz 설치할 때, https://free-eunb.tistory.com/14 에서
# 2. ExecutableNotFound: failed to execute ['dot', '-Tsvg'], make sure the Graphviz executables are on your systems' PATH 에러 
# 이 파트 확인하면서 환경변수 설정 똑같이 따라하면 잘 나옴!

from sklearn.tree import export_graphviz
import pydotplus  # graphviz의 dot language 와의 인터페이스를 제공하는 패키지 임포트
from IPython.display import Image  # IPython의 display와 관련된 Public API

x_list = list(x.columns)
y_list = list(y.drop_duplicates(inplace=False))

# best_dt : 위의 파라미터 실험을 통해 가장 좋은 성능을 낸 파라미터 조합을 가지고 학습한 모델
best_dt = DecisionTreeClassifier(max_depth=7, min_samples_split=3, splitter='best')
best_dt.fit(x_train, y_train)

# export_graphviz() : 의사결정트리에 대한 graphviz dot data를 생성하는 함수
dot_data = export_graphviz(best_dt, out_file=None, feature_names=x_list,
                          class_names=str(y_list), filled=True, rounded=True, special_characters=True)

# matplotlib.rc('font', family='AppleGothic')
graph = pydotplus.graph_from_dot_data(dot_data)  # graphviz의 dot data로부터 트리 그래프 생성

# 트리 그래프를 위한 png 이미지 생성 및 출력
Image(graph.create_png())

# 각 컬럼 별로 의사결정트리 진행
# 각 컬럼 별로 의사결정트리 진행한 과정이 나옴
```

```python
#clf : 돌린 모델
#index : train_x의 칼럼
#sort_values : 값 정렬인데 칼럼명이 0으로 나와서 by=[0]으로 함

feature_scores = pd.DataFrame(best_dt.feature_importances_, 
                           index=x_train.columns).sort_values(by=[0],ascending=False)


#그냥 저장하면 index가 안나와서 뒤에 index 칼럼 추가 
#(위의 46, 2, 30, 8, 40, .... 이 안나오고 1, 2, 3, 4, ....로 진행됨)
feature_scores['index'] = feature_scores.index
feature_scores.to_csv('DT_feature_scores.csv',index = False) 
feature_scores
```

```python
#그래프 그림
import matplotlib.pyplot as plt
import seaborn as sns

f, ax = plt.subplots(figsize=(30, 24))
curve = sns.barplot(x=feature_scores[0], y=feature_scores.index)
curve.set_title("Visualize feature scores of the features")
curve.set_yticklabels(feature_scores.index)
curve.set_xlabel("Feature importance score")
curve.set_ylabel("Features")
plt.show()
```

```python
#0인 값들 제외하고 데이터프레임 생성
print(feature_scores[feature_scores[0] != 0.0]) #0인 값들 제외

DT_importance = list(feature_scores[feature_scores[0] != 0].index)
DT_importance_sort = list(map(int,DT_importance))
DT_importance_sort.sort()
DT_importance = list(map(str,DT_importance_sort))

DT_importance.append('label')

DT_importance_df = data[DT_importance]
DT_importance_df
```
