```python
import numpy as np
import pandas as pd

import torch ## 파이토치, 넘파이의 array를 토치로 받아서 (넘파이는 GPU에 전달을 못해서 토치로 받아서 전달함.)
import torch.nn as nn
```

```python
### pandas, numpy, pytorch의 차이
```

```python
### 데이터 불러오기
from sklearn.datasets import load_iris # colab 환경에서는 데이터가 없기 때문에 sklearn를 통해 데이터를 가져옴.

iris = load_iris()
```

```python
iris # 붓꽃 데이터 - 4개의 열이 많이 들어 있음
```

```python
iris.keys() # 딕셔너리 구조에 포함된 요소를 확인
```

```python
### numpy의 array 자료형 : 모델이 받아들이기 편하게 데이터를 구성해서 보여주는 자료형

iris_data = iris['data'] # 데이터만 따로 추출하여 변수에 할당 # 데이터만 뽑아냄.그래서 하단에 그 지저분한 게 안 뜸.
print(type(iris_data))   # 변수의 자료구조를 확인하니 numpy.ndarray
iris_data # 출력해보면 위에서 본것처럼 array이라는 말이 쓰여져 있고, 단순히 숫자들의 나열에 불과
```

```python
### pandas의 DataFrame 자료형 : 인간이 받아들이기 편하게 데이터를 구성하여 보여줌

iris_colname = iris['feature_names'] # 열 이름을 가져와줌.
iris_colname # 4개의 열이 있는 것을 확인할 수 있음. 
```

```python
### pandas의 dataframe 구조 : 열이름을 통해 해당 숫자들이 어떤 의미를 갖는지 보여줌
# 앞서 numpy의 array 구조가 숫자들의 나열에 불과한 것과 달리 
# pandas의 DataFrame 구조는 
# 열 이름을 통해 해당 숫자들이 어떤 의미를 갖는지 보여줌
iris_pandas = pd.DataFrame(iris_data, columns=iris_colname)
print(type(iris_pandas))
iris_pandas
```

```python
### pytorch : tensor 라는 자료구조를 통해서 데이터를 gpu에 올려서 딥러닝 처리가 가능하게 해줍니다.
# numpy랑 동일하게 둘 다 모델이 처리하기 편하게 데이터를 보여준다.
# 서로 변환이 쉽고 사용되는 문법도 굉장히 비슷하다.

iris_pytorch = torch.tensor(iris_data) # 앞서 할당한 array 구조를 tensor로 변환하여 할당
print(type(iris_pytorch)) # 자료 구조가 torch.Tensor임을 확인 
iris_pytorch # 앞에 붙었던 array라는 말이 사라지고 tensor로 변경
```

```python
torch.tensor(iris_pandas) # pandas의 dataframe 구조는 tensor 변환이 불가능
```

```python
# numpy와 tensor는 서로 변환 되기 때문에 쉽게 변환해주는 메소드(함수)를 가지고 있음

iris_pytorch.numpy()  # tensor를 array 구조로 변경
```

```python
torch.from_numpy(iris_data) # array를 tensor 구조로 변경. 하지만 잘 사용은 안 함.
```
